```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def Pi_MC(N):
    ncirc = 0
    for i in range(N):
        x,y = np.random.rand(2)
        if x**2 + y**2 <= 1:
            ncirc += 1
    return 4*ncirc/N
```

```python
Data_PIMC=[]
for j in range(10000):
    Data_PIMC.append(Pi_MC(10000))
```

```python
plt.plot(Data_PIMC)
plt.ylabel(r'PI_MC', size=20)
plt.show()
```

```python
np.mean(Data_PIMC)
```

```python
np.std(Data_PIMC)
```

```python
np.sqrt(1000)
```

```python
counts, base = np.histogram(Data_PIMC, bins = 100)
```

```python
plt.plot(base[1:], counts, 'k.')
plt.ylabel(r'f(x)', size = 10)
plt.xlabel(r'$x=\pi_{MC}$', size = 10)
plt.show()
```

```python
def Pi_MC2(N):
    X = np.random.rand(N)
    Y = np.random.rand(N)
    return 4*sum(X**2+Y**2<=1)/N
```

```python
Data_PIMC=[]
for j in range(10000):
    Data_PIMC.append(Pi_MC2(10000))
```

```python
import pandas as pd
```

```python
df = pd.read_csv("datasets/dis_us_d.csv")
df.tail()
```

```python
df['Open']
plt.plot(df['Open'])
plt.show()
```

```python
price = np.array(df['Open'])
price
```

```python
Delta = (price[1:]-price[:-1])/price[:-1]
Delta
```

```python
plt.plot(Delta)
plt.show()
```

```python
counts, base = np.histogram(Delta, bins = 100)
```

```python
np.sqrt(len(Delta))
```

```python
plt.plot(base[1:], counts, 'k.')
plt.show()
```

```python
positive = Delta[Delta>0]
positive
```

```python
counts, base = np.histogram(positive, bins = 100)
```

```python
plt.plot(base[1:], counts, 'k.')
plt.show()
```

```python
plt.loglog(base[1:], counts, 'k.')
plt.show()
```

```python
binlist = 10**np.linspace(-2,-1,10)
```

```python
plt.loglog(binlist, binlist, 'k.')
```

```python
binlist = 10**np.linspace(-3,-0.5,100)
counts, base = np.histogram(positive, bins = binlist, density = True)
```

```python
plt.loglog(base[1:], counts, 'k.')
plt.show()
```

```python
negative = Delta[Delta<0]
binlist = 10**np.linspace(-3,-0.5,100)
counts, base = np.histogram(np.abs(negative), bins = binlist, density = True)
```

```python
plt.loglog(base[1:], counts, 'k.')
plt.show()
```
