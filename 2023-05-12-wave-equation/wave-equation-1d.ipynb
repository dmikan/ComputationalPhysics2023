```python
import numpy as np
import matplotlib.pyplot as plt
#from matplotlib import rc
import matplotlib.cm as cm
```

```python
delta_t = 0.001 
t_max = 10
T = np.arange(0,t_max+delta_t,delta_t)

L=1
v=1

delta_x=v*delta_t
X = np.arange(0,L+delta_x,delta_x)
```

```python
A,B = np.meshgrid(T,X)
psi = np.zeros(A.shape)
psi.fill(0)
```

```python
psi[:,0]=np.sin(np.pi*X/L)
```

```python
for i in range(1,X.size-1):
    psi[i][1]=(psi[i+1][0]+psi[i-1][0])/2
```

```python
for n in range(1,T.size-1):
    for i in range(1,X.size-1):
        psi[i][n+1]=psi[i+1][n]+psi[i-1][n]-psi[i][n-1]
```

```python
plt.imshow(psi, extent=(min(T),max(T),min(X),max(X)),aspect=5,origin='lower')
plt.xlabel(r'$t$', size=10)
plt.ylabel(r'$x$', size=10)
plt.colorbar()
plt.show()
```

```python
plt.plot(X,psi)
plt.ylabel(r'$\psi(x,t)$',size=10)
plt.xlabel(r'$x$',size=10)
plt.show()
```
